```python
"""
Q3) Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
Answer-
    Dispersion refers to the way data points are spread out or scattered around a central value (like the mean or median) in a data set. 
    In simple terms, it tells us how "spread out" or "close together" the values are. If the data points are close to the central value, 
    the dispersion is low. If they are spread out, the dispersion is high.

    Measures of Dispersion - Two common ways to measure how data are spread out are variance and standard deviation.
    1. Variance-
        Variance(σ**2) = 1/n ∑(xi - μ) ** 2
        Where:
        xi= each data point
        μ = mean of the data set
        n = number of data points
        Example: Consider the data set: 3, 5, 7
        • The mean is     3 + 5 + 7 / 3 = 15/3 = 5
        • Differences from the mean are: 
          = (3-5), (5-5), (7-5)
          = -2, 0, 2
        • Squared differences:
          = (-2)**2, (0)**2, (2)**2
          = 4,0,4
        • Variance = 4 + 0 + 4 / 3 = 2.67 
        So, the variance is 2.67, showing the spread of data around the mean.


    2. Standard Deviation-
       The standard deviation is the square root of the variance. It tells us how much the values in a data set typically differ from the
       average value. A small standard deviation means that most of the numbers are close to the mean. A large standard deviation means that
       the numbers are more spread out.
       Example: If the variance is 2.67, then the standard deviation is:
       square root of 2.67 is approximately 1.63
       So, the standard deviation is about 1.63, meaning that on average, the data points differ from the mean by about 1.63 units.

    Variance: Measures the average squared distance from the mean.
    Standard Deviation: Measures the average distance from the mean in the same units as the data. 
    Both give us an idea of how spread out the data are, helping us understand whether the values are consistent or vary widely.
"""
```

```python
"""
Q4) What is a box plot, and what can it tell you about the distribution of data?
Answer-
    A box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset that shows the dataset's
    central values, variability, and skewness. 
    Components of a Box Plot:
    1. Box: The main part of the plot (the "box") represents the middle 50% of the data (from Q1 to Q3). The height (or length) of the box 
       shows the interquartile range (IQR), which measures the spread of the middle half of the data.

    2. Median (Q2): A line inside the box represents the median (the middle value of the data set). The median divides the data into two equal
       parts.

    3. Whiskers: The lines extending from the box (called "whiskers") show the range of the data, typically extending to the minimum and 
       maximum values that are within 1.5 times the IQR from the quartiles.

    4. Outliers: Points outside the whiskers are considered outliers. These are unusually high or low values compared to the rest of the data.

    What a Box Plot Tells You About the Data:
    1. Center (Median): The line in the middle of the box shows the median of the data set, giving you an idea of the typical value in the
       data.

    2. Spread (IQR and Range):
       • The length of the box shows the interquartile range (IQR), which is the range of the middle 50% of the data (between Q1 and Q3).
       • The whiskers show the overall spread of the data, excluding outliers.

    3. Skewness: You can see how skewed the data is by looking at the placement of the median line within the box:
       • If the median is centered in the box, the data are symmetric.
       • If the median is closer to the bottom or top, the data are skewed.
       • A long whisker on one side suggests skewness in that direction (longer whisker on the upper side indicates a positive skew, while a
         longer whisker on the lower side suggests negative skew).

    4. Outliers: Points outside the whiskers indicate outliers—data points that are much higher or lower than the rest of the data.
"""
```

```python
"""
Q8) Discuss the conditions under which the binomial distribution is used.
Answer-
    The binomial distribution is used in scenarios where an experiment meets the following conditions:
    1. Fixed Number of Trials: The experiment is repeated a specific number of times (denoted as n). Each repetition is called a "trial."
    
    2. Two Possible Outcomes: Each trial has only two possible outcomes, commonly labeled as "success" and "failure."
    
    3. Constant Probability: The probability of success (denoted as p) remains the same for each trial. The probability of failure is 1 - p.
    
    4. Independent Trials: The outcome of one trial does not affect the outcome of any other trial. Each trial is independent.
    
    Example of Binomial Distribution:
    A classic example is flipping a fair coin 10 times:

    • Fixed number of trials: 10 flips.
    • Two possible outcomes: "Heads" (success) or "Tails" (failure).
    • Constant probability: The probability of getting heads is 0.5 for each flip.
    • Independent trials: The outcome of one flip doesn't affect the others.
    
    In summary, the binomial distribution applies when you have a fixed number of independent trials, each with the same probability of 
    success and two possible outcomes per trial.

"""
```

```python
"""
Q9) Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
Answer-
    Properties of the Normal Distribution:
    The normal distribution, also known as the bell curve or Gaussian distribution, has the following key properties:
    1. Symmetry: It is perfectly symmetrical around the mean. The left and right halves of the curve are mirror images.

    2. Mean, Median, and Mode: In a normal distribution, the mean, median, and mode are all equal and located at the center of the
       distribution.

    3. Bell-shaped Curve: The distribution forms a bell shape, with most data points clustered around the mean and fewer points in the tails.

    4. Asymptotic: The tails of the curve approach but never touch the x-axis. This means extreme values are possible but rare.

    5. Defined by Mean and Standard Deviation: The shape and spread of the curve are determined by two parameters: the mean (µ) and standard 
       deviation (σ). The mean sets the center, and the standard deviation controls the spread (width) of the curve.

    Empirical Rule (68-95-99.7 Rule):
    The empirical rule (or 68-95-99.7 rule) describes how data in a normal distribution is spread out around the mean:
    1. 68% of the data falls within 1 standard deviation (σ) of the mean (µ)
       In notation: μ - 1σ to μ + 1σ

    2. 95% of the data falls within 2 standard deviations of the mean:
       In notation: μ - 2σ to μ + 2σ

    3. 99.7% of the data falls within 3 standard deviations of the mean:
       In notation: μ - 3σ to μ + 3σ

    The normal distribution and empirical rule are essential for many statistical methods, helping to predict data distribution, detect 
    outliers, and make decisions based on probabilities.
"""
```

```python
"""
Q12) Provide an example dataset, calculate both covariance and correlation, and interpret the results.
Answer-
    Example Dataset:
    Hours Studied (X)	Exam Scores (Y)
    2	                   50
    4	                   60
    6	                   70
    8	                   80
    10	                   90

    1. Covariance: Covariance tells us if two variables move together. A positive value means they increase together.
       Covariance = 50
       Interpretation: As hours studied increase, exam scores also increase.

    2. Correlation: Correlation shows the strength of the relationship between two variables. It ranges from -1 to 1. The closer to 1,
       the stronger the positive relationship.
       Correlation = 1
       Interpretation: There's a perfect positive relationship. As hours studied go up, exam scores increase in a perfectly linear way.
"""

import pandas as pd

data = {"Studied_hours" : [2, 4, 6, 8, 10], "Exam_scores": [50, 60, 70, 80, 90]}
df =pd.DataFrame(data)
covariance = df["Studied_hours"].cov(df["Exam_scores"])
correlation = df["Studied_hours"].corr(df["Exam_scores"])

print(df)
print(f"\nCovariance(Studied_hours, Exam_scores): {covariance}")
print(f"Correlation(Studied_hours, Exam_scores): {correlation}")
```
